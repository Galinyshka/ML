#11
Пусть задания с кратким ответом будут задания типа В. Соответственно всего по экзамену вопросов класса В: В
, ... Вк
 Посчитайте процент выполненных и невыполненных заданий по каждому вопросу класса В.

```python
for i in range(len(shortAns)):
  df.insert(len(df.columns), f'B{i+1}', 0)
df
```

```python
for i, _ in df.iterrows():
  shortAns = list(df.loc[i, 'Задания с кратким ответом'])
  for j in range(len(shortAns)):
    df.loc[i, f'B{j+1}'] = shortAns[j]
df
```

```python
cols = df.loc[:, 'B1':'B12'].columns
for col in cols:
  print(df.groupby([col]).agg({col:'count'}) / df.shape[0] * 100)
```

#12
Аналогично и с типом С (ответы с развернутым ответом)

```python
for i in range(int(len(longAns)/4)):
  df.insert(len(df.columns), f'C{i+1}', 0)
df
```

```python
for i, _ in df.iterrows():
  longAns = list(df.loc[i, 'Задания с развёрнутым ответом'])
  for j in range(int(len(longAns)/4)):
    df.loc[i, f'C{j+1}'] = longAns[4*j]
df
```

```python
cols = df.loc[:, 'C1':'C7'].columns
for col in cols:
  print(df.groupby([col]).agg({col:'count'}) / df.shape[0] * 100)
```
